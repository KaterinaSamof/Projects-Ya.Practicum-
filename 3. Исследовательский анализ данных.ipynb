```python
def categorize_floor(data):
    try:
        if data['floor'] == 1:
            return '–ø–µ—Ä–≤—ã–π'
        elif data['floor'] == data['floors_total']:
            return '–ø–æ—Å–ª–µ–¥–Ω–∏–π'
        return '–¥—Ä—É–≥–æ–π'
    except:
        pass
```
